Below is a brief information about each columns of the dataset:

PassengerId: An unique index for passenger rows. It starts from 1 for first row and increments by 1 for every new rows.

Survived: Shows if the passenger survived or not. 1 stands for survived and 0 stands for not survived.

Pclass: Ticket class. 1 stands for First class ticket. 2 stands for Second class ticket. 3 stands for Third class ticket.

Name: Passenger's name. Name also contain title. "Mr" for man. "Mrs" for woman. "Miss" for girl. "Master" for boy.

Sex: Passenger's sex. It's either Male or Female.

Age: Passenger's age. "NaN" values in this column indicates that the age of that particular passenger has not been recorded.

SibSp: Number of siblings or spouses travelling with each passenger.

Parch: Number of parents of children travelling with each passenger.

Ticket: Ticket number.

Fare: How much money the passenger has paid for the travel journey.

Cabin: Cabin number of the passenger. "NaN" values in this column indicates that the cabin number of that particular passenger has not been recorded.

Embarked: Port from where the particular passenger was embarked/boarded.

# # Dealing with missing data

```python
#Checking for missing data
df.isnull().sum()
```

 Age, Cabin and Embarked have 177, 687 and 2 missing values respectively.

Since a large number of values are missing for Cabin, we will not pre-process it and further will reject it while getting features for our machine learning model.

```python
df = df.drop(['Cabin'],axis = 1)
```

```python
# Cabin is dropped thus age and embarked have missing values.
df.isnull().sum()
```

```python
df.columns
```

```python
# Missing values in Age are filled using mean
from sklearn.preprocessing import Imputer
age_imputer = Imputer(missing_values='NaN', strategy='mean', axis=0)
age_imputer = age_imputer.fit(df.iloc[ : , [5]])
df.iloc[:, [5]] = age_imputer.transform(df.iloc[:, [5]])
df.head()
```

```python
# Missing values in age are filled using mean thus only embarked left with missing values.
df.isnull().sum()
```

```python
df['Age']
```

```python
#Missing values in Embarked
df.Embarked.value_counts()
```

We find that category "S" has maximum passengers. Hence, we replace "nan" values with "S".

```python
df['Embarked'] = df['Embarked'].fillna('S')
```

```python
df.Embarked.value_counts()
```

```python
#Check missing data again
df.isnull().sum()
```

All the missing values for relevant features has been filled!

#  Categorical Data

```python
#Encoding Sex
from sklearn.preprocessing import LabelEncoder
labelencoder_sex = LabelEncoder()
df.iloc[:, [4]] = labelencoder_sex.fit_transform(df.iloc[:, [4]])
print df.head() 
```

```python
#Encoding Embarked
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_embarked = LabelEncoder()
df.iloc[:, [10]] = labelencoder_embarked.fit_transform(df.iloc[:, [10]])
print df.head()

df.Embarked.value_counts()
```

```python
#One Hot Encoding Pclass and Embarked
df = pd.get_dummies(df, columns=['Pclass', 'Embarked'])
df.head()
```

# # Creating new features

Combining SibSp & Parch feature, we create a new feature named FamilySize.

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
print df.head()
```

#  Feature Selection

```python
df.head()
```

```python
irrelevant_features = ['PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket']
df = df.drop(irrelevant_features, axis=1)
print df.head()
```

# Selecting Features And Labels

```python
X = df.iloc[:, 1:].values
y = df.iloc[:, 0:1].values
print X.shape, y.shape
```

# Splitting Into Train And Test Set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
print X_train.shape, y_train.shape
```

# Classification and Accuracy

## Logistic Regression

Logistic regression, or logit regression, or logit model is a regression model where the dependent variable (DV) is categorical. This article covers the case of a binary dependent variable—that is, where it can take only two values, "0" and "1", which represent outcomes such as pass/fail, win/lose, alive/dead or healthy/sick. Cases where the dependent variable has more than two outcome categories may be analysed in multinomial logistic regression, or, if the multiple categories are ordered, in ordinal logistic regression.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
clf = LogisticRegression()
clf.fit(X_train, y_train.ravel()) #Trivia : What does ravel do?
y_pred_log_reg = clf.predict(X_test)
acc_log_reg = round( accuracy_score(y_pred_log_reg, y_test) * 100, 2)
print (str(acc_log_reg) + ' %')
```

```python
print y_train.shape
```

```python
print y_train.ravel().shape
```

## Regularisation Constant

```python
clf2 = LogisticRegression(C=0.08)
clf2.fit(X_train, y_train.ravel()) #Trivia : What does ravel do?
y_pred_log_reg2 = clf2.predict(X_test)
acc_log_reg2 = round( accuracy_score(y_pred_log_reg2, y_test) * 100, 2)
print (str(acc_log_reg2) + ' %')
```
